```python
#Import Python Libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action="ignore")
```

```python
#Read csv file
df = pd.read_csv("Salaries.csv")
```

```python
df.head()
```

```python
#Show graphs withint Python notebook
%matplotlib inline
```

```python
#Use matplotlib to draw a histogram of a salary data
_ = plt.hist(df['salary'],bins=10, density= True)
```

```python
#Use seaborn package to draw a histogram
_ = sns.distplot(df['salary']);
```

```python
# Use regular matplotlib function to display a barplot
_= df.groupby(['rank'])['salary'].count().plot(kind='bar')
```

```python
# Use seaborn package to display a barplot
sns.set_style("whitegrid")

ax = sns.barplot(x='rank',y ='salary', data=df, estimator=len)
```

```python
#Scatterplot in seaborn
_ = sns.jointplot(x='service', y='salary', data=df)
```

```python
# side-by-side box plot
_ = sns.boxplot(x='rank',y='salary', data=df, hue='sex')
```

```python
# Pairplot
_ = sns.pairplot(df)
```
